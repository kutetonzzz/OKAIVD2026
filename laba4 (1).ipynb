```python
import pandas as pd
import numpy as np
```

```python
data = "https://raw.githubusercontent.com/EvgeniiSopovGitHub/datasets_uni/refs/heads/main/okaivd/lab_4_raw_data.csv"
df = pd.read_csv(data, sep=';', index_col=0, parse_dates=True)
df.head()
```

```python
df.info()
```

```python
mean_values = df.mean()
df = df.fillna(mean_values)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
unit_cols = [f'unit_{i:02d}' for i in range(1, 11)]
df[unit_cols] = df[unit_cols].fillna(df[unit_cols].mean())
```

```python
df.isnull().sum()
```

```python
df['min'] = df[unit_cols].min(axis=1)
df['median'] = df[unit_cols].median(axis=1)
df['max'] = df[unit_cols].max(axis=1)
```

```python
def mean_std_str(row):
    m = row[unit_cols].mean()
    s = row[unit_cols].std()
    return f"{m:.2f}/{s:.2f}"
df['mean/std'] = df.apply(mean_std_str, axis=1)
```

```python
df.index = df.index.strftime('%Y-%m-%d')
```

```python
numeric_cols = unit_cols + ['min', 'median', 'max']
avg_row = df[numeric_cols].mean()
df.loc['average annual'] = avg_row
df.loc['average annual', 'mean/std'] = np.nan
```
